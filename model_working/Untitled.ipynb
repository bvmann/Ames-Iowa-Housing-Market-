```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
houses=pd.read_csv("../Resources/thirtythree.csv")
```

```python
houses.mean()
```

```python
def label (df):
    df.drop(columns=['MoSold','YrSold'], inplace=True)
   
    le=LabelEncoder()
    
    label_encode = df.select_dtypes(exclude=['number']).columns.tolist()
    

    for x in label_encode:
        col = df[x]
        label = le.fit_transform(col)
        df[x]=label
    le.fit(df['Neighborhood'])
  
      
    return df
```

```python
label(houses)
```
