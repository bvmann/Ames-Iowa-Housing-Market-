## In this notebook I did some feature engineering to cut down on our columns

```python
train = pd.read_csv('../Resources/train_allvalues.csv')
test=pd.read_csv('../Resources/test.csv')
```

```python
train_copy = train.copy()
train_copy.shape
```

```python
#dropped columns first dropped columns that we decided were unneeded or redundant. Then went ahead and found columns 
# with high single value counts and dropped them. 

def feature_selector(df, train=train):
    df.drop(columns=['Id','MiscVal','PoolArea','Utilities','YearRemodAdd','GarageYrBlt'], inplace=True)
    num_values = len(train)
    for x in df.columns: 
        num_unique = df[x].nunique()


        if (num_unique <30):
            v_count= df[x].value_counts()
            single_value = round(v_count.iloc[0] / num_values,2)


            if (single_value > .60) :
                print(f"dropping column : {x}")
                df.drop(columns=x,inplace=True)
                
    return df
```

```python
f_train = feature_selector(train_copy)
```

```python
f_train.shape
```

```python
#combined all porch and deck SQ.Footage and created a column called OutdoorAreaSF. 
#dropped basement and 1st floor and 2nd floor Sq.Footage columns - I did this because these numbers are combined in other 
#pre-existing columns 
def Bsmt_porch (df):
    df = df.drop(columns=['BsmtFinSF2','BsmtFinSF1','BsmtUnfSF','2ndFlrSF','1stFlrSF'])
    df['outdoorAreaSF'] = df['OpenPorchSF']+df['EnclosedPorch']+df['ScreenPorch']+df['WoodDeckSF']
    df['outdoorAreaSF']
    df = df.drop(columns = ['OpenPorchSF','EnclosedPorch','ScreenPorch','WoodDeckSF'])
    

    return(df)
```
